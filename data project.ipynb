```python
# 1. Define the logic for the groups
def categorize_age(age):
    if age < 25:
        return 'Gen Z'
    elif age < 41:
        return 'Millennial'
    elif age < 57:
        return 'Gen X'
    else:
        return 'Boomer'

# 2. Apply this logic to the 'age' column (lowercase!)
df['Age Group'] = df['age'].apply(categorize_age)

# 3. Quick check to confirm it worked
df[['age', 'Age Group']].head()
```

```python
# 1. Define a function to calculate a score based on purchasing history
def calculate_loyalty(row):
    score = 0
    
    # Give points for previous purchases
    if row['previous_purchases'] > 30:
        score += 5
    elif row['previous_purchases'] > 10:
        score += 3
    else:
        score += 1
        
    # Give points for frequency
    if row['frequency_of_purchases'] == 'Weekly':
        score += 5
    elif row['frequency_of_purchases'] == 'Bi-Weekly':
        score += 3
    elif row['frequency_of_purchases'] == 'Monthly':
        score += 2
    else:
        score += 1
        
    return score

# 2. Apply this logic row-by-row (axis=1 means "read across the row")
df['Loyalty Score'] = df.apply(calculate_loyalty, axis=1)

# 3. See who your VIPs are
df[['customer_id', 'previous_purchases', 'frequency_of_purchases', 'Loyalty Score']].head()
```

```python
labels = ['Young Adult', 'Adult','Middle-aged','Senior']
df['age_group'] = pd.qcut(df['age'], q=4 ,labels=labels)
```

```python
frequency_mapping = {
    'Fortnightly ': 14,
    'Weekly ': 7,
    'Monthly':30,
    'Quarterly':90,
    'Bi-Weekly ':14,
    'Annually ': 365,
    'Every 3 Months':90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
import pandas as pd
from sqlalchemy import create_engine
import urllib

# ==========================================
# 1. FILL IN YOUR DETAILS HERE
# ==========================================
server   = 'localhost,1433'      # Try 'localhost' first. If using Docker, might be 'localhost,1433'
database = 'portfolio'         # The name of your database
username = 'sa'             # Your SQL username (often 'sa')
password = 'StrongPass@123'  # The password you type when you open Azure Data Studio
# ==========================================

# 2. DEFINE THE MAC DRIVER
# This is the standard driver name for Mac.
driver = 'ODBC Driver 17 for SQL Server' 

# 3. CREATE THE CONNECTION
try:
    # Build the connection string
    params = urllib.parse.quote_plus(f'DRIVER={driver};SERVER={server};DATABASE={database};UID={username};PWD={password}')
    
    # Connect
    engine = create_engine(f"mssql+pyodbc:///?odbc_connect={params}")
    connection = engine.connect()
    
    print(" SUCCESS! Connected to your database.")
    
except Exception as e:
    print(" CONNECTION FAILED")
    print("Error message:", e)
```

```python
# 1. Define the table name you want to create in the database
table_name = 'customer'

# 2. Write the data to SQL
# 'replace' means: if the table already exists, drop it and upload this new version.
# index=False means: don't upload the row numbers (0, 1, 2...) as a separate column.
df.to_sql(table_name, engine, if_exists='replace', index=False)

print(f" Data successfully saved to the '{table_name}' table in the database!")
```
